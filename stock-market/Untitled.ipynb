```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
path = 'Data/Stocks'
files = os.listdir(path)
dfs = []
for f in files:
    try:
        df = pd.read_csv(os.path.join(path, f))
        df['Source'] = [f[:-4] for _ in range(np.max(df.count()))]
        dfs.append(df)
    except pd.errors.EmptyDataError:
        print(f'can\'t read: {os.path.join(path, f)}')
data = pd.concat(dfs)
print(data.columns)
print(data.shape)
data.head()
```

```python
data = data.sort_values(axis='index', by='Date')
data.head()
```
