```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

## Look for missing data

```python
df.isnull().any()
```

## Check classes balance

```python
(df['output'] == 1).sum() / df['output'].size
```

## Analyse Features

```python
for feature in df.columns:
    
    high = df[feature][df['output'] == 1]
    low = df[feature][df['output'] == 0]

    plt.hist([high, low])
    plt.legend(['high', 'low'])
    plt.xlabel(feature)
    plt.show()
```

```python
plt.figure(figsize=(16, 16))
sns.heatmap(df.corr(), cmap='viridis', annot=True)
plt.show()
```

```python
high = df[['trtbps', 'chol']][df['output'] == 1]
low = df[['trtbps', 'chol']][df['output'] == 0]

plt.scatter(high['trtbps'], high['chol'])
plt.scatter(low['trtbps'], low['chol'])
plt.xlabel('trtbps')
plt.ylabel('chol')
plt.legend(['high', 'low'])
plt.show()
```

## Outliers

```python
# continous = ['age', 'thalachh']
```

```python
fig, axs = plt.subplots(7, 2, figsize=(16, 32))
for i, feature in enumerate(df.columns):
    
    axs[i % 7][i % 2].boxplot(df[feature])
    axs[i % 7][i % 2].set_title(feature)
plt.show()
```
